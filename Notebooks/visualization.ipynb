```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import pandas as pd
```

# Jobs In Demand

```python
#the dataset for the given table is 

job=pd.read_csv('C:/Users/Acer/Job_analysis/new_csv_files/job_in_demand.csv')
job
```

```python

job=job.sort_values('avg_salary',ascending=True)


fig=px.bar(
    job,
    x='avg_salary',
    y='job_title_short',
    orientation='h',
    color='total_jobs',
    hover_data=['min_salary','max_salary'],
    color_continuous_scale='teal'
)

fig.update_layout(
    xaxis_title='Average Salary ($)',
    yaxis_title='Job Role',
    plot_bgcolor='white',
    coloraxis_colorbar=dict(title='Total Jobs')
)



fig.show()
```

```python
#Insights
#Senior data scientist have highest salary but the number of the job is less
# Data Analyst is the role which have the highest number of jobs
# Data Scientist are also in the great demand with the highest salary around 140k
#Business Analyst is the role which have the less jobs as well as less salary
```

# Remote vs Non Remote Salary

```python
df=pd.read_csv('C:/Users/Acer/Job_analysis/new_csv_files/remote_vs_nonremote.csv')
df
```

```python
df = df.sort_values('Prefer', ascending=False)
fig=px.bar(
    df,
    x='Job',
    y=['Remote Pay','Onsite Pay'],
    barmode='group',
    title='Remote vs Non-Remote Salary Comparison',
    color_discrete_map={
        'Remote Pay': '#2E86AB',    
        'Onsite Pay': '#E84855'    
    }
)

fig.update_layout(
    xaxis_title='Job Role',
    yaxis_title='Average Salary ($)',
    legend_title='Work Type',
    xaxis_tickangle=-45,
    plot_bgcolor='white',
    paper_bgcolor='white',
    
)
fig.show()
```

```python
#Insights

#Cloud Engineer have the highest difference in Remote pay than any other
#Senior Data Scientist have the highest remote pay salary
#Business analyst have low remote pay as well as onsite pay
#Data analyst ,Engineer and Senior data analyst have the same number of the salary difference
```

# Salary Percent

```python
percent=pd.read_csv('C:/Users/Acer/Job_analysis/new_csv_files/salary_percent.csv')
percent
```

```python
#so there we need to make the analysis for the job_title_short where the salary is percent 25 percent 50 percent 75 percent 90 earn more

fig = px.bar(
    percent,
    x='job_title_short',
    y=['p25', 'p50', 'p75', 'p90'],
    barmode='group',
    title='Salary Percentile by Job Role',
    color_discrete_sequence=['#2E86AB', '#A23B72', '#F18F01', '#C73E1D']
)

fig.update_layout(
    xaxis_title='Job Role',
    yaxis_title='Salary ($)',
   
    xaxis_tickangle=-45,
    legend_title='Percentile'
)

fig.show()
```

```python
# Senior roles (Data Scientist, Data Engineer) pay the most P90: $200k+
# ML Engineer has the biggest salary gap with experience matters most here
# Data Analyst is the most stable but lowest paying role
# Senior earn more in most of the jobs
```

# Skill Type Analysis

```python
tp=pd.read_csv('C:/Users/Acer/Job_analysis/new_csv_files/skill_type_analysis.csv')
tp
```

```python
tp=tp.sort_values('avg_salary',ascending=True)

fig = px.bar(
    tp,
    x='avg_salary',
    y='skill_type',
    orientation='h',
   
    color='total_jobs',
    title='Skill Type — Demand vs Salary',
    color_continuous_scale='teal',
    hover_data=['total_jobs']
)

fig.update_layout(
    xaxis_title='Average Salary ($)',
    yaxis_title='Skill Type',
    coloraxis_colorbar=dict(title='Total Jobs'),
     
)
fig.update_traces(
    textposition='outside'  
)

fig.show()
```

```python
# Libraries (TensorFlow, PyTorch) pay the most ($139k) AI/ML skills are premium
# Cloud skills (AWS, Azure, GCP) are close second ($133k) with high demand
# Analyst tools (Excel, Tableau) pay the least ($108k) despite decent demand
# Programming has highest demand (47k jobs) but average salary.
```
